```python
from dotenv import load_dotenv
import os
```

```python
load_dotenv(override=True)
```

```python
open_api_key = os.getenv('OPENAI_API_KEY')

if open_api_key:
    print(f"OpenAI API key is exists and begins with {open_api_key[:8]}")
    
else:
    print(f"OpenAI API Key is not set")
```

```python
from openai import OpenAI

openai = OpenAI()
messages = [{"role": "user", "content" : "How to travel to Vietnam by bus from US?"}]
model = "gpt-4o-mini"
response = openai.chat.completions.create(
    model = model,
    messages=messages
)
question = response.choices[0].message.content
print(question)
```

```python
messages = [{"role": "user", "content" : question}]

response_new = openai.chat.completions.create(
    model=model, messages=messages
)

answer = response_new.choices[0].message.content
print(answer)
```
